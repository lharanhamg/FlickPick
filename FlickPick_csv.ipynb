```python
db = pd.read_csv('data/world_imdb_movies_top_movies_per_year.csv')
db
```

```python
db_filtered = db[['id', 'title', 'year', 'duration', 'genre', 'rating_imdb', 'oscar']]
#db_filtered = db_filtered[(db_filtered['year'] >= 1980) & (db_filtered['rating_imdb'] >= 6.8)]
db_filtered = db_filtered.reset_index(drop=True)
db_filtered = db_filtered.dropna()
db_filtered
```

```python
db_filtered.info()
db_filtered.describe()
```

```python
import re
import pandas as pd

def convert_to_minutes(duration_str):
    """
    Converte strings como '1h 38m', '2h', '45m' em minutos inteiros.
    Retorna pd.NA para valores nulos.
    """
    if pd.isna(duration_str):
        return pd.NA

    # Captura horas (ex: '1h', '2h')
    h_match = re.search(r'(\d+)\s*h', duration_str)
    hours = int(h_match.group(1)) if h_match else 0

    # Captura minutos (ex: '38m', '6m')
    m_match = re.search(r'(\d+)\s*m', duration_str)
    minutes = int(m_match.group(1)) if m_match else 0

    return hours * 60 + minutes

# Aplica e mantém dtype que aceita NA
db_filtered['duration'] = (
    db_filtered['duration']
    .apply(convert_to_minutes)
    .astype('Int64')
)
```

```python
fig = px.bar(db_filtered.groupby('year')['title'].count().reset_index(),
             x='year', y='title',
             title='Distribuição de Filmes por Ano (1980-2024)',
             labels={'year': 'Ano', 'title': 'Quantidade de Filmes'})
fig.update_traces(hovertemplate='Ano: %{x}<br>Quantidade de Filmes: %{y}')
fig.show()
```

```python
fig = px.histogram(db_filtered, x="rating_imdb",
                   nbins=int(db_filtered['rating_imdb'].max() - db_filtered['rating_imdb'].min() + 1)*10,
                   title='Distribuição de Filmes por Nota do IMDB',
                   labels={'rating_imdb': 'Nota do IMDB', 'count': 'Quantidade de Filmes'})
fig.update_traces(hovertemplate='Nota do IMDB: %{x}<br>Quantidade de Filmes: %{y}')
fig.update_layout(xaxis={'tickmode': 'linear', 'dtick': 0.1})
fig.show()
```

```python
fig = px.scatter(db_filtered, x="year", y="rating_imdb",
                 color="rating_imdb", size='rating_imdb',
                 title="Relação entre Nota do IMDB, Ano e Quantidade de Filmes",
                 labels={"year": "Ano", "rating_imdb": "Nota do IMDB"},
                 hover_data=['title'],
                 marginal_y="histogram", marginal_x="histogram")

fig.show()
```

```python
#Ordena db_filtered pelo rating_imdb em ordem decrescente
db_filtrados_ordenado = db_filtered.sort_values(by=['rating_imdb'], ascending=False)

#Salva o db ordenado em um arquivo CSV
db_filtrados_ordenado.to_csv('movies2.csv', index=False)
db_filtrados_ordenado
```
